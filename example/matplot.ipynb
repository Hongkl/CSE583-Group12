```python
import pandas as pd
import panel as pn
import folium
import json
from bokeh.plotting import figure
import branca.colormap as cm
import webbrowser
import os
import matplotlib.pyplot as plt  



pn.extension()

state_covid = open("../data/COVID19/COVID19-Rate and Unemployment.csv")
covid_data = pd.read_csv(state_covid)
df=pd.DataFrame(covid_data)
for i in range(len(df)):
    df_covid=pd.DataFrame(df.loc[i, ['March', 'April','May','June','July','August','September']])
    df_unemployment=pd.DataFrame(df.loc[i, ['March_umemployment', 'April_umemployment','May_umemployment','June_umemployment','July_umemployment','August_umemployment','Septembe_umemployment']]) 
    df_unemployment.rename({'March_umemployment': 'March'},axis = 0,inplace = True)
    df_unemployment.rename({'April_umemployment': 'April'},axis = 0,inplace = True)
    df_unemployment.rename({'May_umemployment': 'May'},axis = 0,inplace = True)
    df_unemployment.rename({'June_umemployment': 'June'},axis = 0,inplace = True)
    df_unemployment.rename({'July_umemployment':'July'},axis = 0,inplace = True)
    df_unemployment.rename({'August_umemployment': 'August'},axis = 0,inplace = True)
    df_unemployment.rename({'Septembe_umemployment':'September'},axis = 0,inplace = True)
plt.plot(df_covid, color='green', label='covid rate')
plt.plot(df_unemployment, color='blue', label='unemployment rate')
plt.legend(bbox_to_anchor=(1,1), ncol=3,fancybox=True,shadow=True)
plt.xlabel('Month')
plt.ylabel('Rate %')
plt.title('Yakima County') #the circle only shows the last county right now for example..
```
